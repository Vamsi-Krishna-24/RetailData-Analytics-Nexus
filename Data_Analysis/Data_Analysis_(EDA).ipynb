```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
## All the required libraries are read
```

```python
## The CSV file is read by pandas, named as data.
data = pd.read_csv("/Users/surisettivamsikrishna/Downloads/OnlineRetail.csv",encoding='latin-1')
data
```

```python
#Basic Info 
data.info()

# Summary statistics
data.describe()
```

```python
# Checking all the null values 
data.isnull().sum()
```

```python
# Handling the missing values.
data['Description'].fillna('No_description',inplace=True)
data['CustomerID'].fillna('0',inplace=True)
```

```python
# All the null values has been filled.
data.isnull().sum()
```

```python
data['Quantity'].hist(bins=20,color='skyblue',edgecolor='black')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Quantity')
plt.show()
```

```python
# HeatMap
corr_matrix = data[['Quantity', 'UnitPrice', 'CustomerID']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
